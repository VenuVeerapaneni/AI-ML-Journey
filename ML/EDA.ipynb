```python
import pandas as pd

df = pd.read_csv("train.csv")

print(df.head())

print(df.shape)

print("\n===== DATA TYPES =====")
print(df.dtypes)

print("\n===== DATA INFO =====")
print(df.info())

print("\n===== MISSING VALUES =====")

print(df.isnull().sum())

survival_rate = df["Survived"].mean()

print(
    f"Overall Survival Rate = {survival_rate*100:.2f}%"
)

print(
    df.groupby("Sex")["Survived"]
      .mean()
)

import matplotlib.pyplot as plt

df.groupby("Sex")["Survived"] \
  .mean() \
  .plot(kind="bar")

plt.title("Survival Rate by Gender")
plt.ylabel("Survival Rate")

plt.show()

print(
    df.groupby("Pclass")["Survived"]
      .mean()
)

df.groupby("Pclass")["Survived"] \
  .mean() \
  .plot(kind="bar")

plt.title("Survival Rate by Passenger Class")
plt.ylabel("Survival Rate")

plt.show()

plt.figure(figsize=(8,5))

plt.hist(
    df["Age"].dropna(),
    bins=20
)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")

plt.show()

plt.figure(figsize=(8,5))

plt.hist(
    df["Fare"],
    bins=20
)

plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")

plt.show()

plt.figure(figsize=(8,5))

df.boxplot(
    column="Age",
    by="Survived"
)

plt.title("Age vs Survival")

plt.show()

temp_df = df.copy()

temp_df["Sex"] = temp_df["Sex"].map({
    "male":0,
    "female":1
})

temp_df["Embarked"] = temp_df["Embarked"].map({
    "S":0,
    "C":1,
    "Q":2
})

corr = temp_df.corr(numeric_only=True)

plt.figure(figsize=(10,8))

plt.imshow(corr)

plt.colorbar()

plt.xticks(
    range(len(corr.columns)),
    corr.columns,
    rotation=90
)

plt.yticks(
    range(len(corr.columns)),
    corr.columns
)

plt.title("Correlation Heatmap")

plt.show()


```

===== TITANIC INSIGHTS =====

1. Females had much higher survival rates.

2. First-class passengers survived more often.

3. Fare was strongly related to survival.

4. Most passengers were between 20–40 years old.

5. Passenger class and sex appear highly predictive.
