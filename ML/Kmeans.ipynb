```python
import pandas as pd

data = {
    "Income":[15,16,17,18,80,85,90,95,40,42,45,50],
    "Spending":[39,40,42,45,80,85,88,90,55,58,60,62]
}

df = pd.DataFrame(data)

print(df)

```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters=3,
    random_state=21
)

kmeans.fit(
    df[["Income","Spending"]]
)

df["Cluster"] = kmeans.labels_

print(df)
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df["Income"],
    df["Spending"],
    c=df["Cluster"]
)

plt.xlabel("Income")
plt.ylabel("Spending")

plt.title("K-Means Clusters")

plt.show()

print(kmeans.cluster_centers_)
```

```python
inertia_values = []

for k in range(1,11):

    model = KMeans(
        n_clusters=k,
        random_state=21
    )

    model.fit(
        df[["Income","Spending"]]
    )

    inertia_values.append(
        model.inertia_
    )
```

```python
plt.figure(figsize=(8,5))

plt.plot(
    range(1,11),
    inertia_values,
    marker="o"
)

plt.xlabel("Number of Clusters")

plt.ylabel("Inertia")

plt.title("Elbow Method")

plt.show()
```
